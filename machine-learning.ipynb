## 3.1. Linear Regression

> <span style='font-size: 22px'> Training set </span>

The training set used for model Linear Regression is similar to the training set mentioned in part 2, but to be simple, we will just use 1 feature. <br />
Now, consider the problem of predicting the price of a houses according to its area based on a given dataset. <br />
So, the input of the model is the area of a house and the output is the prediction for its price. Particularly, the format of input and output are :
* _Input_ : a number represents the area of the house.
* _Output_ : a number represents the price of the house.

The training set then looks like :

|                  |   Area   |   Price   | Prediction |
|:----------------:|:--------:|:---------:|:----------:|
| Training example 1 |   60     |  10560    |   10400    |
| Training example 2 |   140    |  20456    |   21030    |
|        ...        |   ...    |    ...    |    ...     |
| Training example m |   210    |  40300    |  359000    |

Now, the task of the model is to predict the price of a house according to its price, which is not in the training set.

> <span style='font-size: 22px'> Model </span>

**Linear Regression model** is used to predict the output from only one input feature.
Each value of the input feature is corresponding to a label, the work of the model is to create a straight line that fits the pattern of every points $\big(x^{(i)}, y^{(i)}\big)$. After training the model, it can predict the output $\hat{y}$ given from a certain input $x$.

<div style='text-align: center;'>
    
</div>

In this model, function “$f$” is a linear function :

$$f(x)=w.x+b$$

Where :
* $w$ is the parameter
* $x$ is the value of the input feature
* $b$ is the bias value

> <span style='font-size: 22px'> Cost Function </span>

For each training example, the input $x$ has corresponding label y and the prediction $\hat{y}$. But the value of $\hat{y}$ may not be equal to y. So, the difference between y and $\hat{y}$ is the “cost”.

The cost of all training examples is represented by the following function called “Cost function” :

$$J=\frac{1}{2m}\sum_{i=1}^{m}\Big(\hat{y}^{(i)}-y^{(i)}\Big)^2=\frac{1}{2m}\sum_{i=1}^{m}\Big(f\big(x^{(i)}\big)-y^{(i)}\Big)^2=\frac{1}{2m}\sum_{i=1}^{m}\big(w.x^{(i)}+b-y^{(i)}\big)^2$$

So, the function $J$ is a quadratic equation with 2 variables $w$ and $b$. Therefore its shape is likely a bowl in 3D space.

> <span style='font-size: 22px'> Gradient Descent </span>

The value of J is the difference between the prediction $\hat{y}$ and the actually value $y$. Hence, we need to minimize the value of J. <br />
To do that, there is an algorithm called "Gradient Descent". This algorithmn just simple adjusts the values of parameters $b$ and $w$ simultaneously. <br />
Particularly, when $J$ is the function of 2 variables $b$ and $w$, a very small step along the $b$-axis is the derivative of $j$ with respect to $w$, and the similar to $b$. So, we can update adjust values of $w$ and $b$ by the derivative of J with respect to themselves respectively. <br />
Repeat the process of continuously updating $w$ and $b$ until the value of $J$ reaches the lowest point on its graph:

$$w=w-\alpha.\frac{d}{dw}J$$

$$b=b-\alpha.\frac{d}{db}J$$

Note that we need to update these simultaneously! <br />
It is worth noticing about the substraction, why it is "-" instead of others?
* If the point $(w, b)$ is at a downward slope, $\frac{d}{dw}J$ and $\frac{d}{db}J$ will be negative, so that point will move forward to the deeper position on the graph of $J$. <br />
* If the point $(w, b)$ is at a upward slope, $\frac{d}{dw}J$ and $\frac{d}{db}J$ will be positive, so that point will move backwards to the deeper position on the graph of $J$. <br />

Finally, after calculating $\frac{d}{dw}J$ and $\frac{d}{db}J$, we have these equations :

$$w=w-\alpha.\frac{1}{m}\sum_{i=1}^{m}\Big(f\big(x^{(i)}\big)-y^{(i)}\Big).x^{(i)}$$

$$b=b-\alpha.\frac{1}{m}\sum_{i=1}^{m}\Big(f\big(x^{(i)}\big)-y^{(i)}\Big)$$

**Step by Step Explanation**
1. For the first time, $w$ and $b$ are initialized with random numbers. So, with these initialized parameters, the value of Cost Function $J$ might be big and not at the minimum. This means the model is very inaccurate.
2. We then use the training set to train the model. First, calculate values $\frac{d}{dw}J$ and $\frac{d}{db}J$. Second, check whether the those values are greater than 0, then update values of $b$ and $w$ respectively.
3. Now, we have better values of $w$ and $b$. Other words, the model is now more accurate. But is might be still not the best. $J$ is at the minimum when $\frac{d}{dw}J$ and $\frac{d}{db}J$ are approximately equal to 0 or very very small.
4. Therefore, we need to repeat step 2 until $\frac{d}{dw}J$ and $\frac{d}{db}J$ are approximately equal to 0 or very very small.

> <span style='font-size: 22px'> Learning rate </span>

Learning rate  is the size of each step. <br />
It is important to consider about the value of $a$ :
* If  is too small, there are too many steps needed for the model to converge. Other words, the training process will be very slow. <br />
* If  is too large, when the point is very close to the convergence point, it will overshoot and can never reach the minimum. Hence, it is likely to fail to converge.

In addition, if the value of $a$ is fixed, it is still able to converge because when the point reach closer to the minimum, the value of $\frac{d}{dw}J$ and $\frac{d}{db}J$ are gradually smaller until they reach 0 at the lowest point.

> <span style='font-size: 22px'> Multiple features </span>

If the input not just contains the area of the house, but the number of floors, number of bedrooms,... the model need to be added some more parameters. In other words, the model now have the vector of parameters $\vec{w}=[w_0, w_1, ..., w_{n-1}, b]$
So, the input is a vector contains multiple values $\vec{x}=[x_0, x_1, ..., x_{n-1}]$. <br />
With mutiple input values, we need to adjust the formula of Model, Cost Function and Gradient Descent :

$$f(\vec{x})=\vec{w}.\vec{x}+b=w_0.x_0+w_1.x_1+...+w_{n-1}.x_{n-1}+b$$

$$J=\frac{1}{2m}\sum_{i=1}^{m}\big(w_j.x_j^{(i)}+b-y^{(i)}\big)^2$$

$$w_j=w_j-\alpha.\frac{1}{m}\sum_{i=1}^{m}\Big(w_j.x_j^{(i)}+b-y^{(i)}\Big)x_j^{(i)}$$

$$b=b-\alpha.\frac{1}{m}\sum_{i=1}^{m}\Big(w_j.x_j^{(i)}+b-y^{(i)}\Big)$$

> <span style='font-size: 22px'> Vectorization </span>

Using for loop to calculate Cost Function and Gradient Descent is slow and complex. Instead, we use Numpy to perform simultaneous calculating. <br />
For example, if we have 2 list of numbers and need to multiply each element of them. With for loop, each multiplication is performed sequentially and added to the result variable step by step. But in Numpy, two list of numbers are stored in "numpy.ndarray" type, we do not need to loop through all elements to multiply each pairs, just `numpy.dot(list_a, list_b)`. <br />
Particularly, instead of using normal for loop like this:
```
a = [1, 2, 3]
b = [4, 5, 6]
result = 0

for i in range(len(a)):
    result += a[i] * b[i]
```
In the this case, $1\times4$, $2\times5$, $3\times6$ are performed sequentially. <br /> <br />
It is much faster and simpler :
```
a = numpy.array([1, 2, 3])
b = numpy.array([4, 5, 6])
result = numpy.dot(a, b)
```
In the this case, $1\times4$, $2\times5$, $3\times6$ are performed at the same time.
